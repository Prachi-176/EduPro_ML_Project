```python
course_summary = transactions.groupby("CourseID").agg(
    Enrollment_Count=("TransactionID", "count"),
    Total_Revenue=("Amount", "sum"),
    Avg_Revenue=("Amount", "mean")
).reset_index()
```

```python
course_summary.head()
```

```python
course_data = pd.merge(
    courses,
    course_summary,
    on="CourseID",
    how="left"
)
```

```python
course_data.shape
```

```python
teacher_course_data = pd.merge(
    transactions[["CourseID", "TeacherID"]],
    teachers[["TeacherID", "YearsOfExperience", "TeacherRating"]],
    on="TeacherID",
    how="left"
)
```

```python
teacher_course_data.head()
```

```python
teacher_course_data.shape
```

```python
teacher_summary = teacher_course_data.groupby("CourseID").agg(
    Avg_Teacher_Experience=("YearsOfExperience", "mean"),
    Avg_Teacher_Rating=("TeacherRating", "mean"),
    Number_of_Teachers=("TeacherID", "nunique")
).reset_index()
```

```python
teacher_summary.head()
```

```python
teacher_summary.shape
```

```python
master_df = pd.merge(
    course_data,
    teacher_summary,
    on="CourseID",
    how="left"
)
```

```python
master_df.head()
```

```python
master_df.shape
```

```python
master_df.columns
```

```python
def price_band(price):
    if price < 500:
        return "Low"
    elif price < 1500:
        return "Medium"
    else:
        return "High"

master_df["Price_Band"] = master_df["CoursePrice"].apply(price_band)
```

```python
master_df[["CoursePrice", "Price_Band"]].head()
```

```python
def duration_bucket(duration):
    if duration < 15:
        return "Short"
    elif duration < 30:
        return "Medium"
    else:
        return "Long"

master_df["Duration_Bucket"] = master_df["CourseDuration"].apply(duration_bucket)
```

```python
master_df[["CourseDuration", "Duration_Bucket"]].head()
```

```python
def rating_tier(rating):
    if rating < 2.5:
        return "Poor"
    elif rating < 3.5:
        return "Average"
    elif rating < 4.5:
        return "Good"
    else:
        return "Excellent"

master_df["Rating_Tier"] = master_df["CourseRating"].apply(rating_tier)
```
